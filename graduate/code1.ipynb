# Data Interaction

```python
# 离散型变量和连续型变量的分割
size = 15
split = 117
data_cont=data_train.iloc[:,split:]
data_cat = data_train.iloc[:,1:117]
cols=data_cont.columns 


# 计算所有连续型变量的Pearson系数，从而确定它们之间的相关性
data_corr = data_cont.corr()

# 系数标准
threshold = 0.5

# 相关性较高的数据组命名
corr_list = []

#寻找相关性比较高的数据组
for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) 

#从高到低排序           
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#打印各个连续变量之间的相关性
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))


# 这将会减少很多模型的训练时间
# cont11和cont12具有强关联性，因此必须去除一个变量，在这里本人选择去除cont11
# cont1和cont9具有较强关联性，因此可以去除一个变量，在这里本人选择去除cont1
# cont6和cont10具有较强关联性，因此可以去除一个变量，在这里本人选择去除cont6
```

```python
for v,i,j in s_corr_list:
    sns.pairplot(data_cont, size=5, x_vars=cols[i],y_vars=cols[j] )
    plt.show()
```

```python
cols = data_cat.columns

#Plot count plot for all attributes in a 29x4 grid
n_cols = 4
n_rows = 29
for i in range(n_rows):
    fig,ax = plt.subplots(nrows=1,ncols=n_cols,sharey=True,figsize=(12, 8))
    for j in range(n_cols):
        sns.countplot(x=cols[i*n_cols+j], data=data_cat, ax=ax[j])
```

```python
from heamy.estimator import Regressor, Classifier
from heamy.pipeline import ModelsPipeline
from sklearn import cross_validation
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error
from heamy.dataset import Dataset

dataset = Dataset(X_train,y_train,X_test)
#创建RF模型和GBDT模型
model_rf = Regressor(dataset=dataset, estimator=RandomForestRegressor, parameters={'n_estimators': 100, 'max_depth':13, 'min_samples_split':100},name='rf')
model_gbdt = Regressor(dataset=dataset, estimator=GradientBoostingRegressor, parameters={'n_estimators': 110, 'max_depth':13, 'min_samples_split': 300}, name='gbdt')
# Blending两个模型
# Returns new dataset with out-of-fold predictions
pipeline = ModelsPipeline(model_rf,model_gbdt)
stack_ds = pipeline.blend(proportion=0.2,seed=111)
#第二层使用LR模型stack
stacker = Regressor(dataset=stack_ds, estimator=LinearRegression)
results = stacker.predict()
# 使用10折交叉验证结果
results = stacker.validate(k=10,scorer=mean_absolute_error)
```
