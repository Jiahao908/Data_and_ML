```python

# 忽略离散型变量
size = 15
split = 117
data_cont=data_train.iloc[:,split:]
data_cat = data_train.iloc[:,1:117]
cols=data_cont.columns 


# 计算所有连续型变量的Pearson系数，从而确定它们之间的相关性
data_corr = data_cont.corr()

# 系数标准
threshold = 0.5

# 相关性较高的数据组命名
corr_list = []

#寻找相关性比较高的数据组
for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) 

#从高到低排序           
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#打印各个连续变量之间的相关性
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))


# 这将会减少很多模型的训练时间
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for v,i,j in s_corr_list:
    sns.pairplot(data_cont, size=5, x_vars=cols[i],y_vars=cols[j] )
    plt.show()
```

```python
from sklearn import preprocessing
data_test = pd.read_csv("test.csv")
le = preprocessing.LabelEncoder()
for col in data_train.columns.values:
    if data_train[col].dtypes == 'object':
        data=data_train[col].append(data_test[col])
        le.fit(data.values)
        data_train[col]=le.transform(data_train[col])
        data_test[col]=le.transform(data_test[col])
```

```python
from sklearn.ensemble import AdaBoostRegressor


def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step

for learning_r in list(frange(0.1,1.1,0.1)):
    for n_esti in list(range(1,21,1)):
        model = AdaBoostRegressor(learning_rate = learning_r, n_estimators = n_esti)
        model.fit(X_train, y_train)
        print(learning_r, n_esti)
        print("{:.2f}".format(model.score(X_train, y_train)))
        print("{:.2f}".format(model.score(X_test, y_test)))
        print("\n")
```

```python
from sklearn.tree import DecisionTreeRegressor


for max_dep in range(1, 20, 1):
    model = DecisionTreeRegressor(max_depth = max_dep , random_state = 0)
    model.fit(X_train, y_train)
    print(max_dep)
    print("{:.2f}".format(model.score(X_train, y_train)))
    print("{:.2f}".format(model.score(X_test, y_test)))
    print("\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.pointplot(max_dep, model.score(X_train, y_train), alpha=0.8, color=)
plt.ylabel('r^2', fontsize=12)
plt.xlabel('max_dep', fontsize=12)
plt.xticks(rotation='vertical') 
plt.show()
```
